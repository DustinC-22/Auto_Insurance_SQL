```python
# Core libraries
import pandas as pd

df = pd.read_csv("Car_Insurance_Claim.csv")
df.head()
```

```python
df.info()
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Check new shape
print("Cleaned dataset shape:", df_cleaned.shape)

# Export customers.csv
customers_cols = ["ID", "AGE", "GENDER", "RACE", "EDUCATION", "INCOME", "CREDIT_SCORE", "MARRIED", "CHILDREN", "POSTAL_CODE"]
df_cleaned[customers_cols].to_csv("customers.csv", index=False)

# Export vehicles.csv
vehicles_cols = ["ID", "VEHICLE_OWNERSHIP", "VEHICLE_YEAR", "VEHICLE_TYPE", "ANNUAL_MILEAGE"]
df_cleaned[vehicles_cols].to_csv("vehicles.csv", index=False)

# Export driving_history.csv
driving_cols = ["ID", "DRIVING_EXPERIENCE", "SPEEDING_VIOLATIONS", "DUIS", "PAST_ACCIDENTS"]
df_cleaned[driving_cols].to_csv("driving_history.csv", index=False)

# Export claims.csv
claims_cols = ["ID", "OUTCOME"]
df_cleaned[claims_cols].to_csv("claims.csv", index=False)
```
